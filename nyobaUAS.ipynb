```python
#CODE-Ngarang

class Pembayaran:
    def __init__(self, id_pembayaran: str, jumlah: float, tanggal: str):
        self.id_pembayaran = id_pembayaran
        self.jumlah = jumlah
        self.tanggal = tanggal
        self.status = False  # Status awal adalah belum dibayar

    def set_status(self, status: bool):
        self.status = status

    def info_pembayaran(self):
        status_str = "Sudah Dibayar" if self.status else "Belum Dibayar"
        return f"ID Pembayaran: {self.id_pembayaran}, Jumlah: {self.jumlah}, Tanggal: {self.tanggal}, Status: {status_str}"


# Struktur Data
daftar_pembayaran = []
pembayaran_dict = {}

def tampilkan_semua_pembayaran():
    if not daftar_pembayaran:
        print("Tidak ada pembayaran yang tersedia.")
    else:
        for pembayaran in daftar_pembayaran:
            print(pembayaran.info_pembayaran())

def cari_pembayaran(id: str):
    if id in pembayaran_dict:
        print(pembayaran_dict[id].info_pembayaran())
    else:
        print(f"Pembayaran dengan ID '{id}' tidak ditemukan.")

def tambah_pembayaran(id_pembayaran: str, jumlah: float, tanggal: str):
    try:
        # Validasi input jumlah
        if not isinstance(jumlah, (int, float)):
            raise ValueError("Jumlah pembayaran harus berupa angka.")

        pembayaran_baru = Pembayaran(id_pembayaran, jumlah, tanggal)
        daftar_pembayaran.append(pembayaran_baru)
        pembayaran_dict[id_pembayaran] = pembayaran_baru
        print(f"Pembayaran dengan ID '{id_pembayaran}' berhasil ditambahkan.")

    except ValueError as e:
        print(f"Kesalahan: {e}")

def ubah_status_pembayaran(id: str, status: bool):
    if id in pembayaran_dict:
        pembayaran_dict[id].set_status(status)
        print(f"Status pembayaran dengan ID '{id}' berhasil diubah.")
    else:
        print(f"Pembayaran dengan ID '{id}' tidak ditemukan.")

def main():
    while True:
        print("\nMenu Manajemen Pembayaran")
        print("1. Tambah Pembayaran")
        print("2. Tampilkan Semua Pembayaran")
        print("3. Cari Pembayaran")
        print("4. Ubah Status Pembayaran")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            id_pembayaran = input("Masukkan ID Pembayaran: ")
            while True:
                try:
                    jumlah = float(input("Masukkan Jumlah Pembayaran: "))
                    break
                except ValueError:
                    print("Jumlah pembayaran harus berupa angka. Silakan coba lagi.")

            tanggal = input("Masukkan Tanggal (YYYY-MM-DD): ")
            tambah_pembayaran(id_pembayaran, jumlah, tanggal)

        elif pilihan == '2':
            tampilkan_semua_pembayaran()

        elif pilihan == '3':
            id_cari = input("Masukkan ID Pembayaran yang dicari: ")
            cari_pembayaran(id_cari)

        elif pilihan == '4':
            id_ubah = input("Masukkan ID Pembayaran yang ingin diubah statusnya: ")
            status_input = input("Apakah sudah dibayar? (y/n): ")
            status = True if status_input.lower() == 'y' else False
            ubah_status_pembayaran(id_ubah, status)

        elif pilihan == '5':
            print("Terima kasih! Program selesai.")
            break

        else:
            print("Pilihan tidak valid. Silakan pilih antara 1-5.")

if __name__ == "__main__":
    main()
```

```python
#CodeKelompok
class Pembayaran: #class pembayaran
#Atribut class
    def __init__(self, id_pembayaran: str, jumlah: float, tanggal: str, status: bool = False):
        self.id_pembayaran = id_pembayaran #id untuk setiap pembayaran
        self.jumlah = jumlah #jumlah belanja dan uang yang dibayarkan
        self.tanggal = tanggal #tanggal pembayaran
        self.status = status #Status pembayaran(default=false)artinya belum dibayar

    def set_status(self, status: bool):
        """Mengubah status pembayaran"""
        self.status = status

    def info_pembayaran(self):
        """Mengembalikan informasi lengkap tentang pembayaran"""
        status_str = "Sudah Dibayar" if self.status else "Belum Dibayar"
        return (f"ID: {self.id_pembayaran}, Jumlah: {self.jumlah}, Tanggal: {self.tanggal}, Status: {status_str}")

#StrukturData
daftar_pembayaran = []  #List semua pembayaran
dict_pembayaran = {}  #Dictionary yang memetakan ID pembayaran ke objek Pembayaran untuk pencarian cepat.

def tampilkan_semua_pembayaran(): #fungsi untuk menampilkan pembayaran
#Fungsi ini mengecek apakah ada pembayaran yang tercatat. Jika tidak ada, akan menampilkan pesan. Jika ada, akan mencetak informasi dari setiap objek
    """Menampilkan semua pembayaran yang ada"""
    if not daftar_pembayaran:
        print("Belum ada pembayaran yang tercatat.")
    else:
        print("Daftar Pembayaran:")
        for pembayaran in daftar_pembayaran:
            print(pembayaran.info_pembayaran())


#Fungsi ini mencari objek Pembayaran berdasarkan ID. Jika ditemukan, akan menampilkan informasi; jika tidak, akan menampilkan pesan kesalahan.
def cari_pembayaran(id_pembayaran: str):
    """Mencari pembayaran berdasarkan ID"""
    try:
        pembayaran = dict_pembayaran[id_pembayaran]
        print("Pembayaran Ditemukan:")
        print(pembayaran.info_pembayaran())
    except KeyError:
        print(f"Pembayaran dengan ID '{id_pembayaran}' tidak ditemukan.")

#Fungsi ini meminta input jumlah dari pengguna dan memastikan bahwa input tersebut adalah angka. Jika input kosong atau bukan angka, pengguna akan diminta untuk mencoba lagi.
def validasi_jumlah(prompt):
    """Validasi input jumlah"""
    while True:
      try:
          nilai = input(prompt)
          if nilai == "":
              print("Input tidak boleh kosong.")
              continue
          else:
              return int(nilai)
      except ValueError:
            print("Invalid input. Input harus berupa angka.")

#Fungsi ini mengumpulkan data untuk membuat objek Pembayaran.
#Memastikan ID unik dan memvalidasi jumlah serta format tanggal sebelum menambahkan objek ke dalam list dan dictionary.
def tambah_pembayaran():
    """Menambahkan pembayaran baru"""
    try:
        id_pembayaran = input("Masukkan ID Pembayaran: ")
        if id_pembayaran in dict_pembayaran:
            print("ID Pembayaran sudah ada. Silakan gunakan ID yang berbeda.")
            return

        jumlah = validasi_jumlah("Masukkan Jumlah Pembayaran: ")

        while True:
            try:
                tanggal = input("Masukkan Tanggal Pembayaran (YYYY-MM-DD): ")
                if '-' not in tanggal:
                    raise ValueError("Format tanggal harus berupa angka dan mengandung tanda '-'!")
                break
            except ValueError as e:
                print(e)

        pembayaran_baru = Pembayaran(id_pembayaran, jumlah, tanggal)

        daftar_pembayaran.append(pembayaran_baru)
        dict_pembayaran[id_pembayaran] = pembayaran_baru
        print("Pembayaran berhasil ditambahkan!")
    except ValueError:
        print("Input jumlah pembayaran harus berupa angka!")

#Fungsi ini mengubah status dari suatu objek Pembayaran menjadi sudah dibayar berdasarkan ID yang diberikan oleh pengguna.
def ubah_status_pembayaran():
    """Mengubah status pembayaran"""
    id_pembayaran = input("Masukkan ID Pembayaran yang akan diubah statusnya: ")
    try:
        pembayaran = dict_pembayaran[id_pembayaran]
        pembayaran.set_status(True)
        print("Status pembayaran berhasil diubah menjadi 'Sudah Dibayar'.")
    except KeyError:
        print(f"Pembayaran dengan ID '{id_pembayaran}' tidak ditemukan.")

#Bagian ini adalah loop utama dari program yang menampilkan menu kepada pengguna dan memproses pilihan mereka.
#Setiap pilihan memanggil fungsi terkait untuk melakukan operasi tertentu.
while True:
    print("\n--- Program Manajemen Pembayaran ---")
    print("1. Tambah Pembayaran")
    print("2. Tampilkan Semua Pembayaran")
    print("3. Cari Pembayaran Berdasarkan ID")
    print("4. Ubah Status Pembayaran")
    print("5. Keluar")

    try:
        pilihan = int(input("Pilih menu (1-5): "))
        if pilihan == 1:
            tambah_pembayaran()
        elif pilihan == 2:
            tampilkan_semua_pembayaran()
        elif pilihan == 3:
            id_pembayaran = input("Masukkan ID Pembayaran: ")
            cari_pembayaran(id_pembayaran)
        elif pilihan == 4:
            ubah_status_pembayaran()
        elif pilihan == 5:
            print("Terima kasih telah menggunakan program ini!")
            break
        else:
            print("Pilihan tidak valid! Silakan masukkan angka antara 1-5.")
    except ValueError:
        print("Input harus berupa angka!")
```
